```python
data = pd.read_csv('IRIS.csv')
data = data.dropna()
data.head()
```

```python
data.drop(['species'], axis=1, inplace = True)
```

```python
sns.pairplot(data)
```

```python
iris_data = data[['petal_length', 'petal_width']]
iris_data.head()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris_data)
```

```python
import scipy.cluster.hierarchy as sch 
dendrogram = sch.dendrogram(sch.linkage(iris_data, method = 'ward')) 
plt.title('Dendrogram')
plt.xlabel('Iris Flower') 
plt.ylabel('Euclidean distances') 
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
groups = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
groups .fit_predict(iris_data)
```

```python
fig = plt.figure(figsize = (9,6))
plt.scatter(iris_data['petal_length'] ,iris_data['petal_width'], c= groups.labels_, cmap='cool')
plt.title(' Iris data divided into three clusters', size=14)
plt.xlabel('Petal length', size=12)
plt.ylabel('Petal width', size=12)
plt.grid()
plt.show()
```
