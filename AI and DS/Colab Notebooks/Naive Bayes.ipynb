```python
data = pd.read_csv("credit_train.csv")
data.head()
```

```python
data.isnull().sum().sum()
```

```python
data=data.dropna()
data.isnull().sum().sum()
```

```python
X = data[['Current Loan Amount','Annual Income','Monthly Debt']]
y = data['Home Ownership'] 
```

```python
data['Home Ownership'].replace(['Own Home','Home Mortgage', 'Rent', 'HaveMortgage'],[1,2,3,4],inplace=True)
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0)
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clf1=GaussianNB()
clf1.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred1=clf1.predict(X_test)
score=accuracy_score(y_pred1,y_test)
print(score)
```

```python
df1=pd.DataFrame(y_test,y_pred1)
df1=pd.DataFrame({"Actual":y_test,'Predicted':y_pred1})
print(df1)
```

```python
feat = np.array([[250000, 600000, 8000]])
print("Predicted: {}".format(clf1.predict(feat)))
print("1 represents Short Own Home, 2 represents Home Mortgage, 3 for Rent, 4 for HaveMortgage")
```

Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
clf2=BernoulliNB()
clf2.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred2=clf2.predict(X_test)
score=accuracy_score(y_pred2,y_test)
print(score)
```

```python
df2=pd.DataFrame(y_test,y_pred2)
df2=pd.DataFrame({"Actual":y_test,'Predicted':y_pred2})
print(df2)
```

```python
feat = np.array([[250000, 600000, 8000]])
print("Predicted: {}".format(clf2.predict(feat)))
print("1 represents Short Own Home, 2 represents Home Mortgage, 3 for Rent, 4 for HaveMortgage")
```

Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
clf3=MultinomialNB()
clf3.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
y_pred3=clf3.predict(X_test)
score=accuracy_score(y_pred3,y_test)
print(score)
```

```python
df3=pd.DataFrame(y_test,y_pred3)
df3=pd.DataFrame({"Actual":y_test,'Predicted':y_pred3})
print(df3)
```
